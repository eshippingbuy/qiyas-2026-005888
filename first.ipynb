```python
paragraphs = "Life is full of challenges, but it is also full of opportunities. Every day brings new chances to learn, grow, and improve ourselves. The way we respond to difficulties often shapes the kind of person we become. With the right mindset, even obstacles can turn into stepping stones.Education plays a key role in this journey. It gives us knowledge, skills, and confidence to face the world. Whether through formal schooling or personal experiences, learning helps us understand ourselves and others better. It opens doors to possibilities we might never have imagined.Friendship is another important part of life. True friends support us in hard times and celebrate with us in good times. They remind us that we are not alone and that sharing joy and sorrow makes life richer. Building strong relationships adds meaning to our daily existence.Hard work and determination are also essential. Success rarely comes overnight; it requires patience and effort. When we stay committed to our goals, we build resilience and strength. Achievements earned through persistence feel more rewarding and lasting.In the end, life is about balance. We need to combine learning, relationships, and effort with rest, joy, and gratitude. By appreciating what we have and striving for what we want, we can create a fulfilling and meaningful life."
```

```python
srt = sorted(paragraphs.lower().split())
print(srt)
```

```python
counts= {}
for i in srt:
    if i not in counts:
        counts[i] =1
    else:
        add = counts[i]+1
        counts[i]= add
print(counts)
```

```python
file = open ("data.txt" ,"a", encoding ="utf-8")
for data in counts:
   file_data=f"{data}\t\t\t{counts[data]}\n"
   file.write(file_data)
```

```python
file = open ("data.txt" ,"r", encoding ="utf-8")
print(file.read())
file.close()
```
