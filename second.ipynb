```python
from bs4 import BeautifulSoup
import requests
```

```python
html_doc = requests.get('https://www.worldometers.info/coronavirus/')
```

```python
soup = BeautifulSoup(html_doc.text, 'html')
```

```python
table_content = soup.find('table')
table_head = table_content.find('thead')
table_body = table_content.find('tbody')
table_row = table_content.find_all('tr')
for data in table_row[8:]:
    # print(table_row)
    # print("---------------------------------------------------------------------------------------")
    pass
```

```python
import csv
with open ("coronavirus_data.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    header_data = []
    for th in table_head.find_all('th'):
        header_data.append(th.text.strip());
    writer.writerow(header_data)
    
    for data in table_row[8:]:
        cells = data.find_all("td")

        row_data = [cell.text.strip() for cell in cells]
        writer.writerow(row_data)

print("Data successfully saved to coronavirus_data.csv!")
```
